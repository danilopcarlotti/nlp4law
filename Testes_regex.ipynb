```python
import pandas as pd
import re

from pathlib import Path
```

```python
df_regex_patterns = pd.read_excel(Path().absolute() / "data/patterns.xlsx")
df_stj = pd.read_excel(Path().absolute() / "data/text_class.xlsx")
```

```python
df_regex_patterns
```

```python
patterns = []
for row in df_regex_patterns.to_dict("records"):
    patterns.append({"name":row["Nome_variável"],"regex":row["Expressão regular"]})
```

```python
new_rows = []
for row in df_stj.to_dict("records"):
    dic_aux = {}
    for dic in patterns:
        found = False
        print(r"{}".format(dic["regex"]))
        if dic["regex"] and re.search(r"{}".format(dic["regex"]), row["text"], re.I):
            found = True
        dic_aux[dic["name"]] = 1 if found else 0
    new_rows.append(dic_aux)
```

```python
new_df = pd.DataFrame(new_rows)
new_df.to_excel(Path().absolute() / "data/resultados_regex.xlsx", index=False)
```
